```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error, accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
import requests
import json
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

print("ğŸš€ STEP 4: ADVANCED ML WITH REAL DATA INTEGRATION")
print("="*60)
print("ğŸ™ï¸ Integrating 4 Cities: Mumbai, Pune, Bangalore, Chennai")
print("ğŸŒ§ï¸ Real weather data + ğŸ¤– ML models + ğŸ Competition analysis")

# Load your simulated data
try:
    df = pd.read_csv('uber_rides_india.csv')
    print(f"âœ… Loaded {len(df)} rides from simulated dataset")
except:
    print("âŒ Please run the data generator first!")
    exit()

# Filter for our 4 focus cities
focus_cities = ['Mumbai', 'Pune', 'Bangalore', 'Chennai']
df = df[df['city'].isin(focus_cities)]
print(f"ğŸ“ Filtered to 4 cities: {len(df)} rides")

# =============================================================================
# PART A: REAL DATA INTEGRATION ğŸŒ
# =============================================================================

print("\nğŸŒ PART A: INTEGRATING REAL DATA")
print("="*40)

def get_real_weather_data(city):
    """
    Simulate real weather API calls (in production, you'd use actual APIs like OpenWeatherMap)
    For demo purposes, we'll create realistic weather patterns based on Indian climate
    """
    # Indian weather patterns by city
    weather_patterns = {
        'Mumbai': {
            'monsoon_months': [6, 7, 8, 9],  # June-September
            'hot_months': [3, 4, 5],         # March-May
            'cool_months': [12, 1, 2],       # December-February
            'base_temp': 28,
            'humidity_factor': 1.3
        },
        'Pune': {
            'monsoon_months': [6, 7, 8, 9],
            'hot_months': [3, 4, 5],
            'cool_months': [12, 1, 2],
            'base_temp': 25,
            'humidity_factor': 1.1
        },
        'Bangalore': {
            'monsoon_months': [6, 7, 8, 9, 10],
            'hot_months': [3, 4, 5],
            'cool_months': [12, 1, 2],
            'base_temp': 23,
            'humidity_factor': 1.0
        },
        'Chennai': {
            'monsoon_months': [10, 11, 12],  # Northeast monsoon
            'hot_months': [4, 5, 6],
            'cool_months': [1, 2, 12],
            'base_temp': 30,
            'humidity_factor': 1.4
        }
    }
    
    # Simulate current month (let's say it's August - monsoon season)
    current_month = 8
    pattern = weather_patterns[city]
    
    # Generate realistic weather based on city and season
    if current_month in pattern['monsoon_months']:
        weather_conditions = ['Heavy_Rain', 'Light_Rain', 'Humid', 'Cloudy']
        weights = [0.3, 0.4, 0.2, 0.1]
        temperature_range = (pattern['base_temp'] - 5, pattern['base_temp'])
        humidity_range = (80, 95)
    elif current_month in pattern['hot_months']:
        weather_conditions = ['Hot', 'Clear', 'Dusty', 'Humid']
        weights = [0.4, 0.3, 0.2, 0.1]
        temperature_range = (pattern['base_temp'] + 3, pattern['base_temp'] + 8)
        humidity_range = (60, 80)
    else:
        weather_conditions = ['Clear', 'Cool', 'Humid', 'Light_Rain']
        weights = [0.5, 0.3, 0.15, 0.05]
        temperature_range = (pattern['base_temp'] - 3, pattern['base_temp'] + 2)
        humidity_range = (65, 85)
    
    return {
        'weather_condition': np.random.choice(weather_conditions, p=weights),
        'temperature': np.random.uniform(temperature_range[0], temperature_range[1]),
        'humidity': np.random.uniform(humidity_range[0], humidity_range[1]),
        'air_quality_index': np.random.randint(50, 300),  # AQI
        'monsoon_intensity': 1.5 if current_month in pattern['monsoon_months'] else 1.0
    }

def get_real_festival_calendar():
    """
    Real Indian festival calendar (simplified)
    In production, you'd use Hindu calendar APIs
    """
    festivals_2024 = {
        'Diwali': {'date': '2024-11-01', 'impact_factor': 2.0, 'duration_days': 5},
        'Holi': {'date': '2024-03-08', 'impact_factor': 1.8, 'duration_days': 2},
        'Eid': {'date': '2024-04-10', 'impact_factor': 1.7, 'duration_days': 3},
        'Dussehra': {'date': '2024-10-12', 'impact_factor': 1.6, 'duration_days': 1},
        'Ganesh_Chaturthi': {'date': '2024-08-31', 'impact_factor': 1.9, 'duration_days': 11},
        'Navratri': {'date': '2024-10-03', 'impact_factor': 1.7, 'duration_days': 9},
    }
    return festivals_2024

def get_real_traffic_patterns(city, hour):
    """
    Simulate real traffic API (Google Maps style)
    Based on actual Indian city traffic patterns
    """
    traffic_patterns = {
        'Mumbai': {
            'peak_morning': [8, 9, 10],
            'peak_evening': [18, 19, 20, 21],
            'base_congestion': 0.7,  # Mumbai always has traffic!
            'peak_multiplier': 2.0
        },
        'Bangalore': {
            'peak_morning': [8, 9, 10],
            'peak_evening': [17, 18, 19, 20],
            'base_congestion': 0.6,
            'peak_multiplier': 1.8
        },
        'Pune': {
            'peak_morning': [8, 9, 10],
            'peak_evening': [18, 19, 20],
            'base_congestion': 0.4,
            'peak_multiplier': 1.5
        },
        'Chennai': {
            'peak_morning': [8, 9, 10],
            'peak_evening': [18, 19, 20, 21],
            'base_congestion': 0.5,
            'peak_multiplier': 1.7
        }
    }
    
    pattern = traffic_patterns[city]
    base_traffic = pattern['base_congestion']
    
    if hour in pattern['peak_morning'] or hour in pattern['peak_evening']:
        traffic_index = base_traffic * pattern['peak_multiplier']
    else:
        traffic_index = base_traffic
    
    # Add randomness
    traffic_index *= np.random.uniform(0.8, 1.2)
    
    return min(traffic_index, 1.0)  # Cap at 100% congestion

# Integrate real data into your dataset
print("ğŸŒ§ï¸ Integrating real weather data...")
for city in focus_cities:
    city_mask = df['city'] == city
    city_weather = get_real_weather_data(city)
    
    df.loc[city_mask, 'real_temperature'] = np.random.normal(city_weather['temperature'], 2, city_mask.sum())
    df.loc[city_mask, 'real_humidity'] = np.random.normal(city_weather['humidity'], 5, city_mask.sum())
    df.loc[city_mask, 'real_aqi'] = np.random.normal(city_weather['air_quality_index'], 30, city_mask.sum())
    df.loc[city_mask, 'monsoon_factor'] = city_weather['monsoon_intensity']

print("ğŸŠ Integrating real festival calendar...")
festivals = get_real_festival_calendar()
# Simulate festival impact (in real project, you'd match actual dates)
df['festival_impact_score'] = df['special_event'].map({
    'Diwali': festivals['Diwali']['impact_factor'],
    'Holi': festivals['Holi']['impact_factor'], 
    'Eid': festivals['Eid']['impact_factor'],
    'Regular': 1.0
})

print("ğŸš— Integrating real traffic data...")
df['real_traffic_index'] = df.apply(lambda row: get_real_traffic_patterns(row['city'], row['hour']), axis=1)

print("âœ… Real data integration complete!")

# =============================================================================
# PART B: MACHINE LEARNING MODELS ğŸ¤–
# =============================================================================

print("\nğŸ¤– PART B: BUILDING ML MODELS")
print("="*35)

# Prepare features for ML
print("ğŸ”§ Feature engineering...")

# Create additional features
df['is_peak_hour'] = df['hour'].isin([8, 9, 18, 19, 20])
df['is_weekend'] = df['day_of_week'].isin([6, 7])
df['is_late_night'] = df['hour'].isin([22, 23, 0, 1])
df['weather_severity'] = df['weather'].map({
    'Clear': 1, 'Hot': 2, 'Humid': 2, 'Dusty': 3, 'Light_Rain': 4, 'Heavy_Rain': 5
})

# Encode categorical variables
le_city = LabelEncoder()
le_segment = LabelEncoder()
le_weather = LabelEncoder()

df['city_encoded'] = le_city.fit_transform(df['city'])
df['segment_encoded'] = le_segment.fit_transform(df['customer_segment'])
df['weather_encoded'] = le_weather.fit_transform(df['weather'])

# Define feature sets
basic_features = ['surge_multiplier', 'base_fare', 'distance_km', 'hour', 'day_of_week']
advanced_features = basic_features + ['real_temperature', 'real_humidity', 'real_aqi', 
                                     'real_traffic_index', 'festival_impact_score']
ml_features = advanced_features + ['city_encoded', 'segment_encoded', 'weather_encoded',
                                  'is_peak_hour', 'is_weekend', 'is_late_night', 'weather_severity']

print(f"ğŸ“Š Total features for ML: {len(ml_features)}")

# =============================================================================
# MODEL 1: PRICE ELASTICITY PREDICTOR
# =============================================================================

print("\nğŸ“ˆ MODEL 1: Price Elasticity Prediction")
print("="*45)

# Prepare data for elasticity model
X_elasticity = df[ml_features]
y_elasticity = df['ride_completed'].astype(int)

X_train, X_test, y_train, y_test = train_test_split(X_elasticity, y_elasticity, test_size=0.2, random_state=42)

# Train Random Forest for elasticity
elasticity_model = RandomForestClassifier(n_estimators=100, random_state=42)
elasticity_model.fit(X_train, y_train)

# Evaluate
y_pred = elasticity_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"âœ… Price Elasticity Model Accuracy: {accuracy:.3f}")

# Feature importance
feature_importance = pd.DataFrame({
    'feature': ml_features,
    'importance': elasticity_model.feature_importances_
}).sort_values('importance', ascending=False)

print("\nğŸ” Top 5 factors affecting ride completion:")
for i, row in feature_importance.head().iterrows():
    print(f"   {row['feature']:20} | {row['importance']:.3f}")

# =============================================================================
# MODEL 2: REVENUE OPTIMIZATION MODEL
# =============================================================================

print("\nğŸ’° MODEL 2: Revenue Optimization")
print("="*38)

# Calculate revenue per request for each surge level
df['revenue_per_request'] = df['final_fare'] * df['ride_completed']

X_revenue = df[ml_features]
y_revenue = df['revenue_per_request']

X_train_rev, X_test_rev, y_train_rev, y_test_rev = train_test_split(X_revenue, y_revenue, test_size=0.2, random_state=42)

# Train Random Forest for revenue prediction
revenue_model = RandomForestRegressor(n_estimators=100, random_state=42)
revenue_model.fit(X_train_rev, y_train_rev)

# Evaluate
y_pred_rev = revenue_model.predict(X_test_rev)
mse = mean_squared_error(y_test_rev, y_pred_rev)
rmse = np.sqrt(mse)

print(f"âœ… Revenue Model RMSE: â‚¹{rmse:.2f}")

# =============================================================================
# MODEL 3: CUSTOMER CHURN PREDICTOR 
# =============================================================================

print("\nğŸƒ MODEL 3: Customer Churn Prediction")
print("="*40)

# Create churn indicator (cancelled due to high surge)
df['churn_risk'] = ((df['surge_multiplier'] >= 2.0) & (df['ride_completed'] == False)).astype(int)

X_churn = df[ml_features]
y_churn = df['churn_risk']

X_train_churn, X_test_churn, y_train_churn, y_test_churn = train_test_split(X_churn, y_churn, test_size=0.2, random_state=42)

# Train Logistic Regression for churn
churn_model = LogisticRegression(random_state=42, max_iter=1000)
churn_model.fit(X_train_churn, y_train_churn)

# Evaluate
y_pred_churn = churn_model.predict(X_test_churn)
churn_accuracy = accuracy_score(y_test_churn, y_pred_churn)

print(f"âœ… Churn Prediction Accuracy: {churn_accuracy:.3f}")

# =============================================================================
# PART C: COMPETITIVE ANALYSIS ğŸ
# =============================================================================

print("\nğŸ PART C: COMPETITIVE ANALYSIS")
print("="*38)

def simulate_competitor_response(uber_surge, city, weather_condition):
    """
    Simulate how Ola responds to Uber pricing
    Based on market research and competitive intelligence
    """
    
    # City-specific competitive dynamics
    city_competition = {
        'Mumbai': {'ola_aggression': 0.9, 'market_share': {'uber': 0.6, 'ola': 0.4}},
        'Bangalore': {'ola_aggression': 0.8, 'market_share': {'uber': 0.55, 'ola': 0.45}},
        'Pune': {'ola_aggression': 0.85, 'market_share': {'uber': 0.65, 'ola': 0.35}},
        'Chennai': {'ola_aggression': 0.9, 'market_share': {'uber': 0.58, 'ola': 0.42}}
    }
    
    city_data = city_competition[city]
    
    # Ola's response strategy
    if uber_surge <= 1.5:
        ola_surge = uber_surge * 0.95  # Slightly cheaper
    elif uber_surge <= 2.0:
        ola_surge = uber_surge * city_data['ola_aggression']  # More aggressive
    else:
        ola_surge = uber_surge * 0.8  # Much more aggressive at high surge
    
    # Weather adjustment (both companies surge less aggressively in bad weather)
    if weather_condition in ['Heavy_Rain', 'Hot']:
        ola_surge *= 1.1  # Less aggressive during bad weather
    
    return {
        'ola_surge': round(ola_surge, 2),
        'uber_market_share': city_data['market_share']['uber'],
        'ola_market_share': city_data['market_share']['ola'],
        'price_difference_pct': round((uber_surge - ola_surge) / ola_surge * 100, 1)
    }

def predict_customer_switching(price_diff_pct, customer_segment):
    """
    Predict probability of customer switching to competitor
    """
    base_switching = {
        'Budget': 0.4,    # High price sensitivity
        'Regular': 0.25,  # Moderate price sensitivity  
        'Premium': 0.1    # Low price sensitivity
    }
    
    base_rate = base_switching[customer_segment]
    
    # Increase switching probability with price difference
    if price_diff_pct > 20:
        switching_prob = min(base_rate * 2, 0.8)
    elif price_diff_pct > 10:
        switching_prob = base_rate * 1.5
    elif price_diff_pct > 5:
        switching_prob = base_rate * 1.2
    else:
        switching_prob = base_rate
    
    return round(switching_prob, 3)

# Apply competitive analysis to dataset
print("ğŸ” Running competitive analysis...")

competitive_data = []
for idx, row in df.iterrows():
    comp_response = simulate_competitor_response(row['surge_multiplier'], row['city'], row['weather'])
    switching_prob = predict_customer_switching(comp_response['price_difference_pct'], row['customer_segment'])
    
    competitive_data.append({
        'ride_id': row['ride_id'],
        'ola_surge': comp_response['ola_surge'], 
        'price_difference_pct': comp_response['price_difference_pct'],
        'switching_probability': switching_prob,
        'market_share_risk': switching_prob * 0.3  # Simplified market share impact
    })

comp_df = pd.DataFrame(competitive_data)
df = df.merge(comp_df, on='ride_id', how='left')

print("âœ… Competitive analysis complete!")

# =============================================================================
# ANALYSIS RESULTS & INSIGHTS
# =============================================================================

print("\nğŸ“Š INTEGRATED ANALYSIS RESULTS")
print("="*40)

# City-wise competitive pressure
city_competition = df.groupby('city').agg({
    'surge_multiplier': 'mean',
    'ola_surge': 'mean',
    'price_difference_pct': 'mean',
    'switching_probability': 'mean',
    'market_share_risk': 'mean'
}).round(2)

print("\nğŸ™ï¸ CITY-WISE COMPETITIVE ANALYSIS:")
print(city_competition)

# Weather impact on competition
weather_competition = df.groupby('weather').agg({
    'surge_multiplier': 'mean',
    'ola_surge': 'mean',
    'switching_probability': 'mean'
}).round(2)

print("\nğŸŒ§ï¸ WEATHER IMPACT ON COMPETITION:")
print(weather_competition)

# ML Model Predictions Summary
print(f"\nğŸ¤– ML MODEL PERFORMANCE SUMMARY:")
print(f"ğŸ“ˆ Price Elasticity Model: {accuracy:.1%} accuracy")
print(f"ğŸ’° Revenue Optimization Model: â‚¹{rmse:.2f} RMSE")  
print(f"ğŸƒ Customer Churn Model: {churn_accuracy:.1%} accuracy")

# Business Recommendations
print(f"\nğŸ’¡ INTEGRATED BUSINESS RECOMMENDATIONS:")
print("="*45)

high_competition_cities = city_competition[city_competition['switching_probability'] > 0.3].index
low_competition_cities = city_competition[city_competition['switching_probability'] <= 0.3].index

print(f"ğŸ”´ HIGH COMPETITION CITIES: {', '.join(high_competition_cities)}")
print(f"   â€¢ Keep surge <1.8x to avoid customer switching")
print(f"   â€¢ Focus on service quality differentiation")

print(f"ğŸŸ¢ LOW COMPETITION CITIES: {', '.join(low_competition_cities)}")  
print(f"   â€¢ Can sustain higher surge (up to 2.5x)")
print(f"   â€¢ Premium pricing strategy viable")

best_weather_for_surge = weather_competition['switching_probability'].idxmin()
print(f"ğŸŒŸ BEST WEATHER FOR SURGE: {best_weather_for_surge}")
print(f"   â€¢ Customers less likely to switch during {best_weather_for_surge.lower()}")

print(f"\nğŸ¯ NEXT STEPS:")
print(f"1. ğŸ“Š Create interactive dashboard")
print(f"2. ğŸ”„ Real-time pricing engine") 
print(f"3. ğŸ“± A/B testing framework")
print(f"4. ğŸ“ˆ Business case presentation")

# Save enhanced dataset
df.to_csv('uber_rides_ml_enhanced.csv', index=False)
print(f"\nğŸ’¾ Saved ML-enhanced dataset: uber_rides_ml_enhanced.csv")
```
