```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Optional: Suppress matplotlib font warnings for emojis
import warnings
warnings.filterwarnings("ignore", message="Glyph.*missing.*DejaVu Sans.*")

plt.style.use('default')
sns.set_palette("Set2")
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
# Optional: use Arial to avoid DejaVu Sans glyph/emoji warnings
plt.rcParams['font.family'] = 'Arial'

print("ðŸ“Š INDIAN MARKET DEEP DIVE ANALYSIS")
print("="*50)
print("ðŸŽ¯ Answering Your Specific Questions:")
print("1. Price increase during demand hours")
print("2. Festival price sensitivity") 
print("3. Weather impact on pricing")
print("4. Competition comparison (Uber vs Ola vs Rapido)")

# Load data
try:
    df = pd.read_csv('uber_rides_india.csv')
    print(f"âœ… Loaded {len(df)} rides for analysis")
except Exception as e:
    print("âŒ Please run the data generator first!")
    print("Error details:", e)
    exit()

# Add competitor pricing simulation
def add_competitor_pricing(df):
    df['ola_base_fare'] = df['base_fare'] * np.random.uniform(0.9, 0.95, len(df))
    df['ola_surge'] = np.where(df['surge_multiplier'] > 2.0,
                               df['surge_multiplier'] * 0.8,
                               df['surge_multiplier'] * 0.9)
    df['ola_final_fare'] = df['ola_base_fare'] * df['ola_surge']
    df['rapido_base_fare'] = df['base_fare'] * np.random.uniform(0.4, 0.6, len(df))
    df['rapido_surge'] = np.where(df['surge_multiplier'] > 1.5, 1.3, 1.1)
    df['rapido_final_fare'] = df['rapido_base_fare'] * df['rapido_surge']
    return df

df = add_competitor_pricing(df)

# Define demand hours
def categorize_hours(hour):
    if hour in [8, 9, 10]:
        return 'Morning Peak'
    elif hour in [18, 19, 20, 21]:
        return 'Evening Peak' 
    elif hour in [22, 23, 0, 1]:
        return 'Late Night'
    elif hour in [12, 13, 14]:
        return 'Afternoon'
    else:
        return 'Regular Hours'
df['demand_category'] = df['hour'].apply(categorize_hours)

# Dashboard setup
fig, axes = plt.subplots(3, 2, figsize=(16, 20))
fig.suptitle('Indian Market Analysis: Demand, Weather & Competition', fontsize=16, y=0.98)

# 1. Price by Demand Hours
demand_analysis = df.groupby('demand_category').agg({
    'base_fare': 'mean',
    'surge_multiplier': 'mean',
    'final_fare': 'mean',
    'ride_completed': 'mean'
}).round(2)
regular_price = demand_analysis.loc['Regular Hours', 'final_fare']
demand_analysis['price_increase_pct'] = (
    (demand_analysis['final_fare'] - regular_price) / regular_price * 100
).round(1)

bars = axes[0,0].bar(demand_analysis.index, demand_analysis['final_fare'], 
                    color=['lightblue', 'orange', 'red', 'yellow', 'lightgreen'], alpha=0.8)
axes[0,0].set_title('Average Fare by Demand Hours', fontsize=14, pad=20)
axes[0,0].set_ylabel('Average Fare (â‚¹)')
axes[0,0].tick_params(axis='x', rotation=45)
for bar, increase in zip(bars, demand_analysis['price_increase_pct']):
    height = bar.get_height()
    label = f'+{increase}%' if increase > 0 else f'{increase}%'
    axes[0,0].text(bar.get_x() + bar.get_width()/2., height + 1,
                   label, ha='center', va='bottom', fontweight='bold', color='red' if increase > 0 else 'black')

# 2. Festival Sensitivity
festival_analysis = df.groupby('special_event').agg({
    'surge_multiplier': 'mean',
    'ride_completed': 'mean',
    'final_fare': 'mean',
    'ride_id': 'count'
}).round(2)
festival_analysis.columns = ['Avg_Surge', 'Completion_Rate', 'Avg_Fare', 'Total_Rides']
festival_data = festival_analysis[festival_analysis.index != 'Regular']

if len(festival_data) > 0:
    x_pos = np.arange(len(festival_data))
    bars1 = axes[0,1].bar(x_pos - 0.2, festival_data['Avg_Surge'], 0.4, label='Avg Surge', color='purple', alpha=0.7)
    bars2 = axes[0,1].bar(x_pos + 0.2, festival_data['Completion_Rate'], 0.4, label='Completion Rate', color='green', alpha=0.7)
    axes[0,1].set_title('Festival Impact: Surge vs Acceptance', fontsize=14, pad=20)
    axes[0,1].set_xlabel('Festival/Event')
    axes[0,1].set_ylabel('Value')
    axes[0,1].set_xticks(x_pos)
    axes[0,1].set_xticklabels(festival_data.index, rotation=45)
    axes[0,1].legend()
    for bar in bars1:
        height = bar.get_height()
        axes[0,1].text(bar.get_x() + bar.get_width()/2., height + 0.02,
                      f'{height:.1f}x', ha='center', va='bottom', fontsize=9)

# 3. Weather Impact
weather_analysis = df.groupby('weather').agg({
    'surge_multiplier': 'mean',
    'ride_completed': 'mean',
    'final_fare': 'mean',
    'wait_time_mins': 'mean'
}).round(2)
weather_colors = {'Clear': 'gold', 'Light_Rain': 'lightblue', 'Heavy_Rain': 'navy', 
                'Hot': 'red', 'Humid': 'orange', 'Dusty': 'brown'}
bars = axes[1,0].bar(weather_analysis.index, weather_analysis['surge_multiplier'],
                    color=[weather_colors.get(w, 'gray') for w in weather_analysis.index],
                    alpha=0.8)
axes[1,0].set_title('Weather Impact on Surge Pricing', fontsize=14, pad=20)
axes[1,0].set_ylabel('Average Surge Multiplier')
axes[1,0].tick_params(axis='x', rotation=45)
for bar, weather in zip(bars, weather_analysis.index):
    height = bar.get_height()
    surge_val = weather_analysis.loc[weather, 'surge_multiplier']
    axes[1,0].text(bar.get_x() + bar.get_width()/2., height + 0.02,
                  f'{surge_val:.2f}x', ha='center', va='bottom', fontweight='bold')

# 4. Competition Comparison (improved annotation)
uber_avg = df['final_fare'].mean()
ola_avg = df['ola_final_fare'].mean()
rapido_avg = df['rapido_final_fare'].mean()
competitors = ['Uber', 'Ola', 'Rapido']
avg_fares = [uber_avg, ola_avg, rapido_avg]
colors = ['black', 'green', 'blue']
bars = axes[1,1].bar(competitors, avg_fares, color=colors, alpha=0.7)
axes[1,1].set_title('Avg. Fare: Uber vs Ola vs Rapido', fontsize=14, pad=20)
axes[1,1].set_ylabel('Average Fare (â‚¹)')
axes[1,1].set_ylim(0, max(avg_fares)*1.20)

for i, (bar, fare) in enumerate(zip(bars, avg_fares)):
    height = bar.get_height()
    axes[1,1].text(bar.get_x() + bar.get_width()/2., height + 5,
                  f'â‚¹{fare:.0f}', ha='center', va='bottom', fontweight='bold')
    if i > 0:
        pct_cheaper = (1 - fare/uber_avg) * 100
        axes[1,1].text(bar.get_x() + bar.get_width()/2., height/2,
                      f'{pct_cheaper:.1f}%\ncheaper', ha='center', va='center',
                      fontweight='bold', color='white', fontsize=10)
axes[1,1].text(1.5, min(avg_fares)*0.2, "Cheaper % compared to Uber", ha='center', color='gray', fontsize=10)

# 5. Hourly Surge Pattern
hourly_stats = df.groupby('hour').agg({
    'surge_multiplier': 'mean',
    'final_fare': 'mean',
    'ride_completed': 'mean'
}).round(2)
axes[2,0].plot(hourly_stats.index, hourly_stats['surge_multiplier'], 'o-', linewidth=3, markersize=6, color='red', label='Surge Multiplier')
axes[2,0].fill_between(hourly_stats.index, hourly_stats['surge_multiplier'], alpha=0.3, color='red')
axes[2,0].set_title('24-Hour Surge Pattern in Indian Cities', fontsize=14, pad=20)
axes[2,0].set_xlabel('Hour of Day')
axes[2,0].set_ylabel('Average Surge Multiplier')
axes[2,0].grid(True, alpha=0.3)
axes[2,0].set_xticks(range(0, 24, 2))
peak_hours = [8, 9, 18, 19, 20, 21]
for hour in peak_hours:
    axes[2,0].axvline(x=hour, color='orange', linestyle='--', alpha=0.7)
axes[2,0].text(9, max(hourly_stats['surge_multiplier'])*0.9, 'Morning\nPeak', 
              ha='center', bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))
axes[2,0].text(19.5, max(hourly_stats['surge_multiplier'])*0.9, 'Evening\nPeak', 
              ha='center', bbox=dict(boxstyle="round,pad=0.3", facecolor="orange", alpha=0.7))

# 6. Price Sensitivity by Customer Segment Heatmap (value labels, proper formatting)
segment_price_sensitivity = df.groupby(['customer_segment', 'surge_multiplier']).agg({
    'ride_completed': 'mean'
}).reset_index()

# Pivot with consistent surge multiplier order (rounded to 1 decimal)
surge_order = sorted(segment_price_sensitivity['surge_multiplier'].unique())
pivot_sensitivity = segment_price_sensitivity.pivot(
    index='customer_segment',
    columns='surge_multiplier',
    values='ride_completed'
)

# Plot improved heatmap
ax = axes[2,1]
mask = pivot_sensitivity.isnull()
sns.heatmap(
    pivot_sensitivity,
    annot=True,
    fmt=".0%",
    cmap="RdYlGn",
    linewidths=0.5,
    cbar_kws={'label': 'Completion Rate'},
    mask=mask,
    ax=ax
)
ax.set_title("Price Sensitivity Heatmap\n(Green = High Acceptance, % of rides completed)", pad=20)
ax.set_xlabel('Surge Multiplier')
ax.set_ylabel('Customer Segment')
ax.set_xticklabels([f"{x:.1f}x" for x in pivot_sensitivity.columns], rotation=45)
ax.set_yticklabels(pivot_sensitivity.index)
plt.tight_layout()
plt.show()
```
